## Data visualization 

## 1. Relation between Rating & Metascore

```python
fig , ax = plt.subplots()

ax.scatter(movie_data['Rating'], movie_data['Metascore'] )

ax.set_title('relation between Rating & Metascore')
ax.set_xlabel('Rating')
ax.set_ylabel('Metascore')
plt.show()

```

Here, we notice that the metascore conforms almost perfectly to the rating; in other words, the higher the rating, the higher the metascore. This confirms the correlation between the rating and the metascore.

## 3. The most popular Genres 

```python
#count how many movies have each Genre
movies_genres=movie_data[['Title','Genre']].groupby('Genre').count() 
# show the first 20 genres 
movies_genres=movies_genres.sort_values(by=['Title'],ascending=False).head(20)

movies_genres.plot(kind='bar')
plt.xlabel('Genre')
plt.ylabel('Genre Count')
plt.title('The most popular Genres')
movies_genres
```

This plot displays the most frequent genres of movies across our data set,It appears that movies of the â€˜Action, adventure , sci-fiâ€™ genres are the most released .

## 4. Revenue By Years :



```python
df_aggregated = movie_data.groupby(movie_data['Year']).agg({'Revenue (Millions)': ['sum']})
df_aggregated
```

```python
import seaborn as sns 

```

```python
sns.lineplot(data=df_aggregated, x='Year', y=df_aggregated['Revenue (Millions)']['sum'], color='#8dbb69')
plt.ylabel('Revenue (Millions)')
plt.title('Revenue By Years')
plt.show()
```

From 2008 to 2010, we notice a continuous increase in revenue, then it decreases slightly between 2010 and 2011, and then it returns to the highest value in 2016.
